```python
import pandas as pd


df = pd.read_csv('/NSE_BANKNIFTY24JANFUT_29th_Dec_to_20th_Jan.csv')
```

```python
df
```

```python
# Task 1: Calculate the exponential moving averages (EMA)
df['ema_10'] = df['close'].ewm(span=10).mean()
df['ema_20'] = df['close'].ewm(span=20).mean()

df.tail(30)
```

```python
df['datetime'] = pd.to_datetime(df['datetime'],format="%Y-%m-%d %H:%M:%S")
df['datetime']
```

```python
df.loc[0,'datetime'].day_name()
```

```python
df
```

```python
df['Days']=df['datetime'].dt.day_name()
df.tail(30)
```

```python
df.set_index('datetime', inplace=True)
```

```python

%matplotlib inline
```

```python
highs=df['high'].resample('D').max()
highs.plot()
```

```python
highs
```

```python
df['current_day_high'] = df.groupby(df.index.date)['high'].transform('max')
```

```python
df.head(30)
```

```python
df['previous_day_high'] = df['current_day_high'].shift(freq='D')
```
