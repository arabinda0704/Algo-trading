```python
import os
import numpy as np
import pandas as pd

# Define a function to calculate EMA
def calculate_ema(series, span):
    return series.ewm(span=span, adjust=False).mean()

# Define the directory path
directory = "/content/drive/MyDrive/algo_trading_data_sets/feb"

# Get a list of all files in the directory
all_files = os.listdir(directory)
symbol_files_ce = [filename for filename in all_files if filename.startswith("NSE_BANKNIFTY24MAR") and filename.endswith("CE.csv")]

# Initialize lists to store buy_prices, sell_prices, sell_types, and profits_or_losses
# buy_prices = []
sell_prices = []
sell_types = []
profits_or_losses = []

# Iterate through each row in the CE DataFrame
for index, row in ce.iterrows():
    symbol_filename = row['symbol'] + ".csv"
    if symbol_filename in symbol_files_ce:
        # Join the filename with the directory path
        file_path = os.path.join(directory, symbol_filename)

        # Load the symbol dataset
        symbol_dataset = pd.read_csv(file_path)
        symbol_dataset['datetime'] = pd.to_datetime(symbol_dataset['datetime'])  # Convert 'datetime' column to datetime64[ns]

        # Calculate ema_20 for the symbol dataset
        symbol_dataset['ema_20'] = calculate_ema(symbol_dataset['close'], span=20)

        # Find the row with the closest datetime value in the symbol dataset
        symbol_row = symbol_dataset[symbol_dataset['datetime'] == row['datetime']].iloc[0]

        # Buy at the close price
        # buy_price = symbol_row['close']
        # buy_prices.append(buy_price)

        # Initialize selling variables
        sell_price = np.nan
        sell_type = np.nan
        profit_or_loss = np.nan

        # Iterate through each row in the symbol dataset starting from the row's datetime
        for _, symbol_row in symbol_dataset[symbol_dataset['datetime'] >= row['datetime']].iterrows():
            # Check conditions for selling
            if (symbol_row['close'] - row['buy_price']) >= 10:
                sell_price = symbol_row['close']
                sell_type = 'point_gain'
                profit_or_loss = sell_price - row['buy_price']
                break  # Exit the loop if condition met
            elif symbol_row['close'] < symbol_row['ema_20']:
                sell_price = symbol_row['close']
                sell_type = '1min_ema20'
                profit_or_loss = sell_price - row['buy_price']
                break  # Exit the loop if condition met
            elif (symbol_row['datetime'] - row['datetime']).seconds >= 15 * 60:
                sell_price = symbol_row['close']
                sell_type = 'threshold'
                profit_or_loss = sell_price - row['buy_price']
                break  # Exit the loop if condition met

        # Append selling details to lists
        sell_prices.append(sell_price)
        sell_types.append(sell_type)
        profits_or_losses.append(profit_or_loss)
    else:
        # buy_prices.append(np.nan)
        sell_prices.append(np.nan)
        sell_types.append(np.nan)
        profits_or_losses.append(np.nan)

# Add the buy_prices, sell_prices, sell_types, and profits_or_losses lists to the CE DataFrame
# ce['buy_price'] = buy_prices
ce['sell_price'] = sell_prices
ce['sell_type'] = sell_types
ce['profit_or_loss'] = profits_or_losses
```
